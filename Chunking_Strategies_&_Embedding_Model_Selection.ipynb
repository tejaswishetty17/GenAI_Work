```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_core.documents import Document
```

```python
#Step 1: Sample long documnet (could be from a loader)

text = """
Langchain is a framework for developing application powered by language models.
It enables applications that are:
- Data-aware: connect a language model to other sources of data
- Agentic: allow a language model to other interact with its environment

###Features

Langchain provides modular abstraction for:
1.Prompt templates
2.LLMs
3.Agents
4.Tools
5.Memory
6.Indexing and Retrieval

##Use Cases

-Documnet Q&A (RAG)
-Chatbots
-Agents
-Data Agumented Generation
-Summarization
"""

#Convert into Langchain Document
doc = Document(page_content=text)
docs = [doc]
docs
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size = 100,
    chunk_overlap=50,
    separators=["\n\n", "\n", ".", " ", ""]
)

chunks = splitter.split_documents(docs)
```

```python
for i, chunk in enumerate(chunks, 1):
  print(f"\n-- Chunk {i} --\n{chunk.page_content}")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import TextLoader

# load text file
loader = TextLoader("/content/sample_data/langchain_doc.txt")
docs = loader.load()

splitter = RecursiveCharacterTextSplitter(chunk_size=800, chunk_overlap=50, separators=["\n\n", " ", ""])
chunks = splitter.split_documents(docs)
```

```python
print(len(chunks))
```

```python
import os
os.environ['OPENAI_API_KEY'] = "<OPEN_AI_TOKEN>"
```

```python
from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain_community.vectorstores import Chroma

openai_embeddings = OpenAIEmbeddings(model="text-embedding-3-small")

#Create a vector store for OpenAI embeddings
openai_vectorstore = Chroma.from_documents(documents=chunks, embedding=openai_embeddings, collection_name="openai_store1")
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

#Use BGE model
hf_embeddings = HuggingFaceEmbeddings(model_name="BAAI/bge-base-en-v1.5")

#create a vector store for HuggingFace embeddings
hf_vectorstore = Chroma.from_documents(documents=chunks, embedding=hf_embeddings, collection_name="hf_store1")
```

```python
query = "Which vector stores langchain supports?"

print("\n OpenAI Embeddings Results:")
openai_results = openai_vectorstore.similarity_search(query, k=2)
for i, doc in enumerate(openai_results, 1):
  print(f"\n[OpenAI Result {i}]\n{doc.page_content[:300]}")

print("\n HuggingFace BGE Results:")
hf_results = hf_vectorstore.similarity_search(query, k=2)
for i, doc in enumerate(hf_results, 1):
  print(f"\n[HF Result {i}]\n{doc.page_content[:300]}")
```

```python
#Create a vector store for OpenAI embeddings in persist directory and not in memory
openai_vectorstore = Chroma.from_documents(
    documents=chunks,
    embedding=openai_embeddings,
    persist_directory="openai_store_dir1",
    collection_name="openai_store"
    )
```

```python
from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain_community.vectorstores import Chroma

openai_embeddings = OpenAIEmbeddings(model="text-embedding-3-small")

#Load persisted vectorstore
openai_vectorstore = Chroma(
    persist_directory="openai_store_dir1",
    embedding_function=openai_embeddings,
    collection_name="openai_store"
)
```

```python
query = "Which vector stores langchain supports?"

print("\n OpenAI Embeddings Results:")
openai_results = openai_vectorstore.similarity_search(query, k=2)
for i, doc in enumerate(openai_results, 1):
  print(f"\n[OpenAI Result {i}]\n{doc.page_content[:300]}")
```
